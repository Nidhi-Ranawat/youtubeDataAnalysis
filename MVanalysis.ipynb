## Visualisation

```python
import seaborn as sns
# Convert the 'publishedAt' column to datetime
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

# Plot views according to date for each artist
plt.figure(figsize=(12, 6))
sns.lineplot(x='publishedAt', y='views', hue='artist', data=df)
plt.xlabel('Date')
plt.ylabel('Views')
plt.title('Views by Published Date for Each Singer/artist')
plt.xticks(rotation=45)
# plt.ticklabel_format(style='plain')
#  "get current axes"
plt.gca().ticklabel_format(style='plain', axis='y')
plt.legend()
plt.show()
```

```python
# most_viewed_videos = df.groupby('artist')['views']
most_viewed_videos = df.groupby('artist')['views'].idxmax()
```

```python
# np.array(most_viewed_videos)
most_viewed_videos
most_viewed_df = df.loc[most_viewed_videos, ['artist', 'vid_name', 'views']]
```

```python
top_ten_viewed = df.sort_values('views', ascending=False, inplace=False).iloc[:10,:]
top_ten_viewed.set_index(pd.Index(range(1, 11)), inplace=True)
top_ten_viewed.loc[:,['vid_name', 'artist','views']].rename(columns={'vid_name': 'TOP 10 MOST VIEWD VIDEOS'})
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="artist",y="subs",data=df)
plt.xlabel('Channelartist')
plt.ylabel('Subscriberscount')
plt.gca().ticklabel_format(style="plain", axis='y')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="views",y="likes",data=df,hue=df['artist'])
plt.gca().ticklabel_format(style="plain")
plt.title("Views vs. Likes")
plt.show()
```
